### Parte 1: Fundamentos de bases de datos
1. Fundamentos de bases de datos y para ciencia de datos
Las bases de datos son estructuras abstractas para almacenar datos ya sean estructurados o no estructurados (depende del uso de caso). Al momento de escoger una base de datos es necesario tener en cuenta ciertos criterios que pueden facilitar o hacer mas dificil su uso como lo son, una base de datos relacional o no relacional, tiempo de respuesta de los accesos a los registros (cache), el nivel de disponibilidad que se espera, el nivel de consistencia, entre otros. Estos aspectos son los que pueden hacer una diferencia en el rendimiento. A su vez, tambien es necesario tener un buen diseño y comunicacion con la base de datos.

2. Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos
Un data warehouse es una entidad central de informacion la cual se conforma de diferentes fuentes de informacion ya sean bases de datos relacionales, no relacionales, sistemas de transacciones, etc. la cual despues pueden ser accesadas de esta entidad central para mejores analisis de informacion, realizar decisiones informadas, analisis de datos historicos, consistencia de datos, precision y calidad, entre otros. La importancia para ciencias de datos recae en la cantidad de informacion de diferentes fuentes, la organizacion de los datos y la capacidad de realizar las acciones previamente mencionadas con el data warehouse lo cual facilita una de las fases del trabajo de analisis en la ciencia de datos.
