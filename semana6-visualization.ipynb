6. Escala los datos, si consideras necesario

````python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_nona)
X_scaled
````

````python
X_scaled.shape
````

7. Reduce las dimensiones con PCA, si consideras necesario.
    1. Indica la varianza de los datos explicada por cada componente seleccionado. Para actividades de exploración de los datos la varianza > 70%
    2. Indica la importancia de las variables en cada componente

````python
from sklearn.decomposition import PCA
pca = PCA(0.95)
X_pca = pca.fit_transform(X_scaled)
X_pca.shape
````

Con los datos escalados y corridos por el algoritmo de PCA se puede observar el siguiente arreglo con el mismo numero de componentes que calculo el PCA pero indicando la varianza por cada componente que se calculo. Nada mas los primeros 4 componentes son capaces de explicar el 50% del dataset original. A si mismo, podemos observar el aporte de cada componente en la varianza total hasta llegar al nivel de varianza que se indico cuando se inicia el algoritmo.

````python
pca.explained_variance_ratio_
````

````python
pca.n_components_
````

````python
X_pca
````

8. Elabora los histogramas de los atributos para visualizar su distribución

````python
PC_components = np.arange(pca.n_components_) + 1

_ = sns.set(style='whitegrid', font_scale=1.2)

fig, ax = plt.subplots(figsize=(10,7))

_ = sns.barplot(x = PC_components, y = pca.explained_variance_ratio_*100, color='b')

plt.xlabel('N-th Principal Component')
plt.ylabel('Variance explained')
plt.show()
````

````python
def myplot(score,coeff,labels=None):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    scalex = 1.0/(xs.max() - xs.min())
    scaley = 1.0/(ys.max() - ys.min())
    plt.scatter(xs * scalex,ys * scaley, c = df_nona['Y'])
    for i in range(n):
        plt.arrow(0, 0, coeff[i,0], coeff[i,1],color = 'r',alpha = 0.5)
        if labels is None:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var"+str(i+1), color = 'g', ha = 'center', va = 'center')
        else:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i], color = 'g', ha = 'center', va = 'center')
    plt.xlim(-1,1)
    plt.ylim(-1,1)
    plt.xlabel("PC{}".format(1))
    plt.ylabel("PC{}".format(2))
    plt.grid()
````

Este grafico es para demostrar los 'scores' de las observaciones en los componentes principales del PCA y utiliza vectores para representar los coeficientes de las variables en los componentes. Dicho en otras palabras, permite tambien obtener mas observaciones sobre la variacion de las variables de entrada y la separabilidad de las clases, en este caso solo dos uno o cero, en relacion con los componentes principales. Las flechas, su angulo en especifico representan la contribucion que tiene ese particular componente en la direcion que va hacia. La longitud de la flecha representa que tanto peso tiene esa contribucion.

````python
myplot(X_pca[:,0:2], np.transpose(pca.components_[0:2, :]))
plt.show()
````
